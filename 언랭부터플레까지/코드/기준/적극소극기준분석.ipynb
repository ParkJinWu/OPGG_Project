```
def get_gameIds(nick):
  #accountId를 뽑기위해 매개변수로 받아온 닉네임을 사용하여 뽑음
  summoner = "https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/" + nick +'?api_key=' + api_key
  s = requests.get(summoner)

  accountId = s.json()['accountId']
  #뽑은 accountId로 매치리스트 100개를 뽑음
  matchlist = "https://kr.api.riotgames.com//lol/match/v4/matchlists/by-account/" + accountId +'?queue=420&api_key=' + api_key
  ml = requests.get(matchlist)

  #100개의 매치리스트 중 가장 최근에 플레이 한 10개의 게임을 분석
  matchData = pd.DataFrame(ml.json())
  matches = pd.DataFrame(dict(matchData['matches'])).T    #게임 아이디를 뽑기위해 matches컬럼을 다시 데이터프레임으로 만듬
  matches = matches['gameId']   #matches컬럼의 데이터프레임에서 gameId컬럼만 뽑음
  gameIds = []
  for i in range(10):
    gameIds.append(matches[i])    #최근 10개의 게임아이디 뽑음

  return gameIds
```

```
def get_gameData(gameIds):
  gameData = pd.DataFrame()
  for i in range(len(gameIds)):    
    
    match='https://kr.api.riotgames.com/lol/match/v4/matches/' + str(gameIds[i]) + '?api_key=' + api_key  #뽑은 게임아이디로 각각의 인게임데이터를 뽑음
    r = requests.get(match)

    if r.status_code == 200: # response가 정상
        pass

    elif r.status_code == 429: #Rate limit exceeded
        while True: # 429error가 끝날 때까지 무한 루프
            if r.status_code == 429:

                print('try 10 second wait time')
                time.sleep(10)

                r = requests.get(match)
                print(r.status_code)

            elif r.status_code == 200: #다시 정상이면 무한루프 중지
                print('recovery api cost')
                break

    elif r.status_code == 503: #Service unavailable
        print('service available error')

        while True:
            if r.status_code == 503 or r.status_code == 429:

                print('try 10 second wait time')
                time.sleep(10)

                r = requests.get(match)
                print(r.status_code)

            elif r.status_code == 200: # 다시 정상이면 무한루프 중지
                print('recovery api cost')
                break
    elif r.status_code == 403: # Forbidden(api갱신)
        print('you need api renewal')
        print('break')
        break

    # 위의 예외처리 코드를 거쳐서 내려왔을 때 해당 코드가 실행
    game = pd.DataFrame(list(r.json().values()), index=list(r.json().keys())).T
    gameData = pd.concat([gameData,game])    
    #game에 하나의 게임 데이터를 뽑아 gameData에 붙임, 10번 반복해 10개의 게임 데이터를 gameData에 집어 넣음

  return gameData
```

```
def get_inGameData(gameData):
  part_list = list(gameData['participants']) #participants컬럼을 list로 변환
  name_list = list(gameData['participantIdentities'])
  match_df = pd.DataFrame()
  name_df =  pd.DataFrame()
  for i in range(len(part_list)):
    for j in range(len(part_list)):
        match = pd.DataFrame(list(part_list[i][j].values()),index = list(part_list[i][j].keys())).T
        match_df = match_df.append(match)

        name = pd.DataFrame(list(name_list[i][j].values()),index = list(name_list[i][j].keys())).T
        name_df = name_df.append(name)
   
  match_df.index = range(len(match_df)) 
  
  name1_df = pd.DataFrame(list(name_df['player']))
  stats_list = list(match_df['stats'])
  inGameData = pd.DataFrame(stats_list)
  inGameData['summonerName'] = name1_df['summonerName']

  return inGameData
```

```
def get_aveKIR(nick,inGameData):
  i = 0
  j = 0
  index = 0
  idx = []
  KA = []
  isWin = []

  while(i<=99):
    if inGameData['summonerName'][i] == nick:
      index = i;
    if i % 10 == 9:
      KA.append(inGameData['kills'][index] + inGameData['assists'][index])
      isWin.append(inGameData['win'][index])
      idx.append(index)
    i+=1

  a = 0
  b = 10
  tk = 0
  i = 0
  j = 0
  totalKill = []
  for i in range(len(isWin)):
    for j in range(a,b):
      if inGameData['win'][j] == isWin[i]:
        tk += inGameData['kills'][j]
    totalKill.append(tk)
    tk = 0
    a = b
    b += 10 

  killInvolvementRate = []

  for i in range(len(totalKill)):
    killInvolvementRate.append(int(KA[i] / totalKill[i] * 100))

  sum_KIV = 0
  for i in range(len(killInvolvementRate)):
    sum_KIV += killInvolvementRate[i]
  
  ave_KIV = sum_KIV / len(killInvolvementRate)

  return ave_KIV
```
